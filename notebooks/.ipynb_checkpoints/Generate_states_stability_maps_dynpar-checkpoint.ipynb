# Generating dynamic states of parcellations

This notebook shows how to the `dynpar` library to generate dynamic states of parcellations using functional MRI at the individual level.  


```python
# loading the used libraries
import os
import nibabel as nb
import matplotlib.pyplot as plt
import warnings
from nilearn import plotting, datasets
from nilearn.masking import unmask
warnings.filterwarnings("ignore", category=UserWarning)

# Loading the dynamic states of parcellation library
import dynpar
```

Important parameters 
* `path_to_data` (str): folder path of fmri data (e.g. `/home/amal/nilearn_data/nyu_rest/`) for all scans.
* `idx_first_session` (int): index of the first fMRI scan. 
* `idx_last_session` (int): index of the last session of fMRI scan.
* `smoothing_fwhm` (int): FWHM smoothing kernel
* `regress_confounds` (bool, default=False): Confounds regression
* `n_clusters` (int): number of clusters
* `n_replications` (int): number of replicated seed-based kmeans parcellation for each sliding window
* `max_iter` (int): Maximum iterations of the kmeans clustering algorithm
* `window_length` (int): Number of timepoints in a sliding window
* `shift` (int): Number of timepoints shifted between two consecutive sliding windows
* `cluster_size_threshold` (float): The cluster size probability over the total number of sliding windows (0.1 means 10%)
* `min_dice_threshold` (float): The minimal required distance between two seed-based parcellations to be clustered in the same state.
* `seeds_coords` (list): specify the seed coordinates index (e.g., medoids[32] is the index of the PCC seed)

### Download and loads the NYU resting-state test-retest dataset
See documentation: <a href="https://nilearn.github.io/modules/generated/nilearn.datasets.fetch_nyu_rest.html#nilearn.datasets.fetch_nyu_rest">nilearn.datasets.fetch_nyu_rest</a>

```python
# Downloads 3 sessions from one subject 
nyu_dataset = datasets.fetch_nyu_rest(n_subjects=1, sessions=[1,2,3], data_dir=None, resume=True, verbose=1)
path_to_data, tail = os.path.split(nyu_dataset.func[0])
path_to_data = os.path.join(os.environ['HOME'], 'nilearn_data/nyu_rest/')
```

### Specify all the parameters

```python
pattern_nii = "*session*/*/func/*.nii.gz"

idx_first_session = 0

idx_last_session = 1

smoothing_fwhm=6

regress_confounds = False

n_clusters = 12

n_replications = 1

max_iter=50

window_length= 60

shift= 30

cluster_size_threshold = 0.1

min_dice_threshold = 0.3

networks_labels = ['PCC']

black_bg= True

fmri = 'rest'
```

Loading an instance `funcMRI` class `FuncMRIManagement`. This class includes functional MRI data utility methods (e.g., loading time series and loading the mask) 

```python
funcMRI = dynpar.FuncMRIManagement() 
```

Loading a group parcellation from the `MIST` atlas used as a mask for fMRI data

```python
parcellations = datasets.fetch_atlas_basc_multiscale_2015(version='sym')

networks_122 = parcellations['scale122']

mask_img = funcMRI.load_mask(networks_122)
```

### Loading time series from sessions by concatenation as a Dask array

<p>Dask arrays coordinate many Numpy arrays, arranged into chunks within a grid. They support a large subset of the Numpy API.</p>
<a href="https://mybinder.org/v2/gh/dask/dask-examples/master?urlpath=lab/tree/array.ipynb">Dask array examples</a> 

```python
darr_tseries = funcMRI.load_tseries(path_to_data, mask_img, idx_first_session=idx_first_session, regress_confounds=regress_confounds, pattern_nii=pattern_nii,
                              idx_last_session=idx_last_session, smoothing_fwhm=smoothing_fwhm)
```

```python
darr_tseries
```

### Instanciating the dynamic parcels class `dynpar`
We implemeted the `dynpar` library to generate dynamic states of parcellations

```python
# Create instance of the dynamic_parcellation class
dynp_instance = dynpar.dynamic_parcellation()
```

```python
# Identify `medoids` for the regions within the group parcellation as seed networks

medoids, mni_medoids = dynp_instance.get_seeds(networks_122)


seeds_coords = [medoids[32]]
```

```python
chunksize_voxels = darr_tseries.chunksize[0]
```

Our subject-specific dynamic parcellation approach consisted on five main steps:

<ol>
<li>we generate multiple <b>short time window parcels</b> for the whole cerebral cortex per subject.</li>
<li>Then, we identify the parcel associated with a particular <b>seed-voxel</b>. </li>
<li>We calculate the <b>pairwise similarity of these seed-based parcellations</b> based on the <b>Dice similarity score</b>.</li>
<li>we apply a <b>Hierarchical clustering</b> on this similarity matrix to group these parcellations into <b>a set of clusters (or dynamic states of parcellations)</b> according to a threshold imposed on the Dice similarity matrix, as well as the number of sliding windows included in the state. </li>
<li>For each cluster and each seed, we average of all its seed-based parcellations to obtain the final dynamic state stability maps. </li>
</ol>

Generating states stability maps using the `generate_stability_for_dynamic_brain_parcels` method of the instantiated class 
* Returns `l_darr_stab_maps`: list of Dask arrays, 
 State Stability maps for a given seed (n_states, n_voxels)

```python
l_darr_stab_maps = dynp_instance.generate_stability_for_dynamic_brain_parcels(darr_tseries, mask_img,
                                                          seeds_coords, n_replications, n_clusters, 
                                                          cluster_size_threshold = cluster_size_threshold,
                                                          min_dice_threshold=min_dice_threshold, 
                                                          chunksize_voxels=chunksize_voxels,                                               
                                                          window_length=window_length, shift=shift) 
```

### Show the state stability maps for a given seed 

```python
cmap_name = ["Blues", "Reds", "Greens", "Purples", "hot"]
threshold = 0.5
annotate = True
vmax=1
display_mode='x'
cut_coords=[-44, -34, -10, -6, 0, 16, 46] #PCC, x
resampling_interpolation ='continuous' 
subject = ''
```

```python
for darr_stab_maps_medoid, network_label in zip(l_darr_stab_maps, networks_labels):
    
    n_states = darr_stab_maps_medoid.shape[0]
    
    for idx in range(n_states):
        
        state_img = unmask(darr_stab_maps_medoid[idx, :], mask_img, order='F')
        
        cmap = plt.cm.get_cmap(cmap_name[idx])
        
        title = "Dynamic state " + str(idx+1) + "-" + fmri + "-" + network_label + "-" + subject 
        
        plotting.plot_stat_map(state_img, threshold=threshold, black_bg=black_bg,
                               annotate=annotate, cmap=cmap, vmax=vmax,
                               display_mode=display_mode, cut_coords=cut_coords,
                               title=title, resampling_interpolation=resampling_interpolation)
                              
        plt.show()
```
